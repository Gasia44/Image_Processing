```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xp = [0, 4, 8, 12, 16, 20, 24, 28, 32 ]
fp = [0, 8, 0, -8, 0, 8,    0, -8, 0]
np.interp(2.5, xp, fp)
s_2 =np.interp(np.arange(33), xp, fp)
```

```python
xp = [0, 8, 16, 24, 32 ]
fp = [0, 8, 0, -8, 0]
np.interp(2.5, xp, fp)
s_1 =np.interp(np.arange(33), xp, fp)
```

```python
xp = [0, 4,  8, 12, 16, 20, 24, 28, 32 ]
fp = [8, 0, -8,  0, 8,  0, -8,  0,  8 ]
np.interp(2.5, xp, fp)
c_2 =np.interp(np.arange(33), xp, fp)
```

```python
xp = [0, 8, 16, 24, 32 ]
fp = [8, 0, -8,  0, 8 ]
np.interp(2.5, xp, fp)
c_1 =np.interp(np.arange(33), xp, fp)
```

```python
plt.figure(figsize = (20,4))
plt.grid()
plt.plot(s_2, label='s2(x)', linestyle='--', linewidth=4)
plt.plot(c_2, label='c2(x)', linestyle='--', linewidth=2)
plt.plot(s_1, label='s1(x)',  linewidth=4)
plt.plot(c_1, label='c2(x)',  linewidth=2)
plt.xticks(np.arange(34))
plt.legend(prop={'size': 14})
plt.savefig('Task_2_Q', dpi=100)
```
